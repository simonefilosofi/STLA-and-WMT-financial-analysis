## 1. Introduction to Cost of Capital

The cost of capital represents the minimum return that a company must earn on its investments to satisfy its investors (both debt holders and equity holders). It serves as a critical benchmark for evaluating new projects and as a key metric for understanding a company's financial performance. In this analysis, we examine and compare the cost of capital for Walmart and Stellantis, providing insights into their risk profiles, financing costs, and investment hurdle rates.

## 2. Key Metrics and Parameters

The analysis utilizes the following key metrics and market parameters:

| Parameter | Value | Description |
|-----------|-------|-------------|
| Risk-Free Rate | 4.27% | Based on current Treasury yields |
| Market Risk Premium | 5.5% | Expected market return above the risk-free rate |
| Corporate Tax Rate | 23.4% | Effective corporate tax rate |

### 2.1 Company-Specific Metrics

| Metric | Walmart (WMT) | Stellantis (STLA) |
|--------|---------------|-------------------|
| Beta | 0.70 | 1.49 |
| Pre-tax Cost of Debt | 5.8% | 8.8% |
| Equity Ratio (E/V) | 92.91% | 50.38% |
| Debt Ratio (D/V) | 7.09% | 49.62% |

## 3. Cost of Capital Components Analysis

### 3.1 Cost of Equity

The cost of equity represents the return required by shareholders for their investment in the company. Using the Capital Asset Pricing Model (CAPM):

Cost of Equity = Risk-Free Rate + Beta  Market Risk Premium

#### Walmart:
- Cost of Equity = 4.27% + (0.70  5.5%) = **8.10%**

#### Stellantis:
- Cost of Equity = 4.27% + (1.49  5.5%) = **12.47%**

This significant difference in the cost of equity (8.10% vs. 12.47%) primarily stems from the companies' different beta values, which measure the volatility of their stock prices relative to the overall market:

- **Walmart's beta of 0.70** indicates lower volatility than the market, reflecting its defensive retail business model and stable cash flows. This translates to a lower risk premium demanded by equity investors.

- **Stellantis's beta of 1.49** demonstrates significantly higher volatility than the market, consistent with the cyclical nature of the automotive industry. This higher beta results in a substantially higher risk premium required by shareholders.

### 3.2 Cost of Debt

The cost of debt represents the effective interest rate a company pays on its debt obligations, adjusted for the tax shield provided by interest expense deductibility.

#### Walmart:
- Pre-tax Cost of Debt: 5.8%
- After-tax Cost of Debt = 5.8%  (1 - 23.4%) = **4.44%**

#### Stellantis:
- Pre-tax Cost of Debt: 8.8%
- After-tax Cost of Debt = 8.8%  (1 - 23.4%) = **6.74%**

The 3% difference in pre-tax borrowing costs (5.8% vs. 8.8%) reflects:

1. **Credit Quality Differences**: Walmart's stronger credit rating allows it to borrow at more favorable rates.
2. **Financial Risk Perception**: Stellantis's higher leverage ratio likely contributes to its higher borrowing costs.
3. **Industry Risk Factors**: The automotive industry typically faces higher borrowing costs than the retail sector due to its cyclical nature and capital-intensive operations.

### 3.3 Weighted Average Cost of Capital (WACC)

The WACC represents the overall cost of capital for the company, weighted by the relative proportions of debt and equity in its capital structure:

WACC = (E/V  Cost of Equity) + (D/V  After-tax Cost of Debt)

#### Walmart:
- WACC = (92.91%  8.10%) + (7.09%  4.44%) = **7.84%**

#### Stellantis:
- WACC = (50.38%  12.47%) + (49.62%  6.74%) = **9.63%**

## 1. Introduction to Dividend Policy

A company's dividend policy represents its strategy for distributing earnings to shareholders versus retaining them for reinvestment. This policy reflects management's perspective on shareholder value creation, business maturity, financial flexibility needs, and future growth opportunities. In this analysis, we compare the dividend policies of Walmart and Stellantis, examining their approaches to shareholder returns and what these policies reveal about their business strategies and financial health.

## 3. Analysis of Dividend Policies

### 3.1 Walmart's Dividend Strategy

Walmart's dividend policy reveals a company that balances shareholder returns with significant reinvestment for growth:

- **Low Current Yield, High Growth Potential**: With a dividend yield of just 0.96%, Walmart offers shareholders a modest current income. However, the estimated dividend growth rate of 13.10% suggests strong potential for future dividend increases.

- **Balanced Payout Approach**: The 38.80% payout ratio indicates that Walmart is returning a significant portion of earnings to shareholders while still retaining the majority (61.20%) for reinvestment.

- **High ROE Driving Growth**: Walmart's impressive 21.41% ROE demonstrates excellent capital efficiency, allowing the company to generate substantial returns from its retained earnings. This high ROE enables Walmart to maintain both dividend increases and business growth from the same earnings pool.

- **Growth-Oriented Valuation**: Walmart's P/E ratio of 40.52 reflects investors' willingness to pay a premium for expected future growth. This high multiple suggests the market values Walmart's reinvestment of retained earnings and expects these investments to generate substantial future returns.

- **Sustainable Dividend Growth**: With expected earnings growth of 4.38% and a calculated dividend growth potential of 13.10%, Walmart appears positioned to continue its dividend growth trajectory, though the actual dividend growth rate may be lower than the calculated maximum.

### 3.2 Stellantis's Dividend Strategy

Stellantis employs a distinctly different dividend approach that prioritizes current income over growth:

- **High Current Yield**: With a 5.90% dividend yield, Stellantis offers investors significant immediate income, positioning itself as an income-oriented investment. This yield is approximately six times higher than Walmart's.

- **Similar Payout Structure**: Despite the vastly different yields, Stellantis's 40.77% payout ratio is remarkably similar to Walmart's, indicating that both companies allocate comparable portions of their earnings to dividends.

- **Low Capital Efficiency**: Stellantis's 6.72% ROE indicates significantly lower capital efficiency compared to Walmart. This lower ROE constrains the company's ability to generate growth from retained earnings.

- **Value-Oriented Valuation**: The P/E ratio of 6.91 suggests that investors view Stellantis as a value stock rather than a growth opportunity. This low multiple may reflect concerns about future profitability or industry challenges.

- **Challenged Growth Prospects**: The expected earnings decline of 8.70% contrasts sharply with the calculated potential dividend growth rate of 3.98%, suggesting potential pressure on future dividend sustainability if earnings continue to decline.

## 4. Comparative Analysis and Strategic Implications

### 4.1 Divergent Shareholder Return Models

The dividend policies of Walmart and Stellantis represent two classic but divergent approaches to delivering shareholder returns:

- **Walmart: Growth-Oriented Model** focuses on total shareholder return through a combination of modest current dividends and substantial capital appreciation potential. The company's low dividend yield but high estimated growth rate fits the profile of a growth-oriented dividend payer.

- **Stellantis: Income-Oriented Model** emphasizes immediate cash returns to shareholders through high dividend yields. The company's high dividend yield but limited growth prospects positions it as an income stock.

### 4.2 Dividend Sustainability Analysis

The sustainability of each company's dividend policy depends on several factors:

#### Walmart:
- **Strong Fundamentals**: The combination of high ROE, moderate payout ratio, and positive earnings growth provides a solid foundation for sustainable dividend growth.
  
- **Growth-Reinvestment Balance**: Walmart's retention of 61.20% of earnings allows significant reinvestment in strategic initiatives like e-commerce expansion, supply chain optimization, and store renovations.
  
- **Conservative Payout Ratio**: The moderate payout ratio provides a buffer against earnings volatility, reducing the risk of dividend cuts during challenging economic periods.

#### Stellantis:
- **Earnings Pressure**: The projected 8.70% decline in earnings raises concerns about long-term dividend sustainability at current levels if negative growth trends persist.
  
- **Industry Challenges**: As an automotive manufacturer facing industry disruption (electrification, autonomous driving), Stellantis may need more capital for reinvestment than its current retention ratio allows.
  
- **Cyclical Exposure**: The automotive industry's cyclicality creates inherent volatility in earnings, which could threaten dividend stability during downturns despite the similar payout ratio to Walmart.

### 4.3 Investor Implications

The contrasting dividend policies create different investment propositions:

- **Walmart Appeals To**:
  - Growth-oriented investors seeking total return
  - Dividend growth investors with long time horizons
  - Investors willing to accept lower current income for potential future increases
  - Those seeking defensive exposure with dividend upside

- **Stellantis Appeals To**:
  - Income-focused investors prioritizing current yield
  - Value investors attracted to low P/E multiples
  - Investors who believe the company's challenges are temporary or overstated
  - Those willing to accept higher risk for higher current income

## 5. Dividend Policy in Industry Context

### 5.1 Retail Industry Perspective (Walmart)

Walmart's dividend approach aligns with evolving retail industry dynamics:

- **Digital Transformation Requirements**: Significant capital investments are required to compete in the evolving omnichannel retail landscape, justifying Walmart's higher retention ratio.
  
- **Defensive Sector Positioning**: As a consumer staples retailer, Walmart benefits from relatively stable cash flows that support dividend consistency even during economic downturns.
  
- **Scale Advantages**: Walmart's industry-leading scale enables it to maintain competitive pricing while still generating returns sufficient to support both growth investments and dividends.

### 5.2 Automotive Industry Perspective (Stellantis)

Stellantis's high-yield approach reflects automotive industry characteristics and challenges:

- **Mature Industry Dynamics**: The traditional automotive industry's limited organic growth opportunities often lead to higher dividend payout strategies to attract investors.
  
- **Cyclical Considerations**: High current yields can compensate investors for the inherent cyclicality and volatility of automotive stocks.
  
- **Transition Challenges**: The automotive industry's expensive transition to electric vehicles creates competing capital demands that may eventually pressure dividend sustainability.
  
- **Consolidation Benefits**: As a recently merged entity, Stellantis may be using its high dividend yield to reward shareholders while working through post-merger synergies and optimizations.

## 6. Future Outlook and Considerations

### 6.1 Walmart's Dividend Future

Several factors support the continuation of Walmart's measured, growth-oriented dividend approach:

- The significant gap between estimated dividend growth potential (13.10%) and projected earnings growth (4.38%) provides flexibility for dividend increases above earnings growth rates, at least temporarily.

Stellantis faces more significant dividend policy challenges:

The substantial gap between the current 5.90% yield and the 10-year Treasury yield (approximately 4.3%) suggests investors are demanding a significant risk premium, reflecting concerns about the sustainability of both the business model and the dividend.

## 7. Conclusion

Walmart and Stellantis illustrate fundamentally different approaches to dividend policy, representing classic growth-oriented versus income-oriented models. These differences extend beyond simple payout mechanics to reflect each company's industry position, financial health, and strategic priorities.

Walmart's lower-yield, higher-growth dividend strategy reflects its strong competitive position, significant reinvestment opportunities, and consistent financial performance. ## 1. Introduction to Growth and Value Investing

The distinction between growth and value stocks represents one of the fundamental paradigms in investment theory and portfolio construction. Growth investors focus on companies with strong revenue and earnings expansion potential, often willing to pay premium prices for future prospects. Value investors, by contrast, seek companies trading below their intrinsic worth, focusing on current fundamentals rather than future expectations. This analysis examines Walmart and Stellantis through the growth-value lens, highlighting how these companies exemplify different investment styles.

## 2. Classification Methodology

Our classification framework assesses companies based on three critical financial metrics that traditionally distinguish growth stocks from value stocks:

| Metric | Growth Indicator | Value Indicator |
|--------|-----------------|-----------------|
| Price-to-Earnings (P/E) Ratio | High (>20) | Low (20) |
| Dividend Yield | Low (<2%) | High (2%) |
| Earnings Growth Rate | High (>8%) | Low (8%) |

Based on these criteria, companies receive a score in each category, with the predominant characteristic determining their classification as either growth, value, or hybrid.
